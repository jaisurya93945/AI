## imports

```python
import pandas as pd
import logging
import sys

#logging handlers
for handler in logging.root.handlers[:]:
    logging.root.removeHandler(handler)

logging.basicConfig(
    level=logging.INFO,
    format="%(asctime)s [%(levelname)s] %(message)s",
    stream=sys.stdout
)

logging.info("Logging initialized successfully.")
```

## DatasetLoader Class

```python
class DatasetLoader:
  def __init__(self, filepath):
    self.filepath = filepath
    logging.info(f"Initializing DatasetLoader with filepath: {filepath}")

  def load_data(self):
    try:
      data = pd.read_csv(self.filepath)
      logging.info("File loaded successfully.")
      return data
    except FileNotFoundError:
      logging.error("File not found.")
      return None

  def get_shape(self, data):
    if data is not None:
      logging.info(f"Dataset shape: {data.shape}")
      return data.shape
    else:
      logging.warning("Data is None.")
      return None
```

```python
loader = DatasetLoader("/content/retail_customer_segmentation.csv")
data = loader.load_data()
loader.get_shape(data)
```

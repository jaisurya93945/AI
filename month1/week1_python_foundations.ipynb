## DatasetLoader Class

```python
class DatasetLoader:
  def __init__(self, filepath):
    self.filepath = filepath
    logging.info(f"Initializing DatasetLoader with filepath: {filepath}")

  def load_data(self):
    try:
      data = pd.read_csv(self.filepath)
      logging.info("File loaded successfully.")
      return data
    except FileNotFoundError:
      logging.error("File not found.")
      return None

  def get_shape(self, data):
    if data is not None:
      logging.info(f"Dataset shape: {data.shape}")
      return data.shape
    else:
      logging.warning("Data is None.")
      return None
```

```python
loader = DatasetLoader("/content/retail_customer_segmentation.csv")
data = loader.load_data()
loader.get_shape(data)
```

## DAY - 2
### Matrix Multiplication and Dot product and ML Prediction Formula

## Dot Product

### example

#### prediction = w1*x1 + w2*x2 + w3*x3

```python
import numpy as np

a = np.array([1,2,3])
b = np.array([1,2,3])

print("Vector a:", a)
print("Vector b:", b)

result = a @ b

print("Dot product:", result)
```

## what it means

### [1,2,3] • [1,2,3]

###

### Calculation:

#### 1×1 + 2×2 + 3×3
#### = 1 + 4 + 9
#### = 14 #which is output

```python
weights = np.array([2,3,4])
features = np.array([1,2,3])

prediction = weights @ features

print("Prediction:", prediction,"shapes are:",weights.shape,features.shape)
```

```python
weights = np.array([[2,3,4],[5,6,7]])
features = np.array([1,2,3])

prediction = weights @ features
print("shapes are:",weights.shape,features.shape)
print("Prediction:", prediction)
```

## Matrix Multiplication

### Matrix Multiplication Rule

#### (m × n) @ (n × p) = (m × p)

```python
X = np.array([[1,2,3],
              [4,5,6]])

W = np.array([[1],[2],[3]])

print("X shape:", X.shape)
print("W shape:", W.shape)

y = X @ W

print("Predictions:")
print(y)
```
